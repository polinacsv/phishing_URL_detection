```python
# === Setup ===
import os, numpy as np, pandas as pd
from pathlib import Path
from collections import Counter

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import precision_recall_curve, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import make_scorer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV

# XGBoost (optional; will skip if not installed)
try:
    from xgboost import XGBClassifier
    HAS_XGB = True
except Exception:
    HAS_XGB = False

# TensorFlow CNN (optional)
USE_CNN = False
try:
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
    HAS_TF = True
except Exception:
    HAS_TF = False
    USE_CNN = False

# === Load data ===
df = pd.read_csv("../data/df_combined.csv")

# Your selected columns
columns_to_keep = [
    'url','label',
    'num_digits','url_length','num_subdomains','num_path_segments',
    'num_suspicious_keywords','url_entropy','jaccard_hostname_path','path_self_similarity',
    'is_in_alexa','log_ranking','suspicious_subdomain_or_path',
    'domain_word2vec_similarity','avg_token_length','digit_letter_ratio'
]
df = df[columns_to_keep].copy()

# Train/test split (stratified)
X = df.drop(columns=['label','url'])
y = df['label'].astype(int)
X_train, X_test, y_train, y_test, url_train, url_test = train_test_split(
    X, y, df['url'], test_size=0.2, random_state=42, stratify=y
)

# === Helper: threshold selection with FN penalty ===
def pick_threshold_min_cost(y_true, y_prob, fn_weight=5.0):
    """
    Choose threshold that minimizes cost = fn_weight*FN + 1*FP.
    Returns best_threshold, dict(cost, FN, FP)
    """
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    # Convert PR points to thresholds; last point has no threshold -> handle separately
    candidates = np.r_[thr, 1.0 - 1e-12]
    best = (None, np.inf, None, None)
    for t in candidates:
        yhat = (y_prob >= t).astype(int)
        FN = int(((y_true==1) & (yhat==0)).sum())
        FP = int(((y_true==0) & (yhat==1)).sum())
        cost = fn_weight*FN + FP
        if cost < best[1]:
            best = (t, cost, FN, FP)
    return float(best[0]), {"cost": int(best[1]), "FN": int(best[2]), "FP": int(best[3])}

def evaluate_at_threshold(y_true, y_prob, thr):
    y_pred = (y_prob >= thr).astype(int)
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, zero_division=0),
        "Recall": recall_score(y_true, y_pred, zero_division=0),
        "F1 Score": f1_score(y_true, y_pred, zero_division=0),
    }

# === Models & grids ===
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = []

def add_result(name, y_true, y_prob, meta=None, fn_weight=5.0):
    thr, diag = pick_threshold_min_cost(y_true, y_prob, fn_weight=fn_weight)
    met = evaluate_at_threshold(y_true, y_prob, thr)
    row = {"Model": name, "Threshold": thr, **met, "FN_weight": fn_weight}
    if meta: row.update(meta)
    results.append(row)

# Common scaler
scaler = StandardScaler()

# 1) Logistic Regression L2
pipe_l2 = Pipeline([
    ("scaler", scaler),
    ("clf", LogisticRegression(max_iter=5000, solver="lbfgs", class_weight=None))
])
grid_l2 = {"clf__C":[0.01,0.1,1,10,100]}
gs_l2 = GridSearchCV(pipe_l2, grid_l2, cv=cv, scoring="recall", n_jobs=-1)
gs_l2.fit(X_train, y_train)
prob_l2 = gs_l2.best_estimator_.predict_proba(X_test)[:,1]
add_result("LogReg (L2)", y_test, prob_l2, {"BestParams": gs_l2.best_params_})
```

```python
# ==== Robust char-level CNN (dynamic alphabet, consistent mapping) ====
import pandas as pd, numpy as np, string
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, precision_recall_curve
from tensorflow import keras
from tensorflow.keras import layers

# --- Load & split ---
df = pd.read_csv("../data/df_combined.csv", usecols=["url","label"])
urls = df["url"].astype(str)
y = df["label"].astype(int)

url_train, url_test, y_train, y_test = train_test_split(
    urls, y, test_size=0.2, random_state=42, stratify=y
)

# --- Build alphabet dynamically from TRAIN URLs (lowercased, ASCII only) ---
max_len = 200
train_slice = url_train.str.lower().str.slice(0, max_len)
unique_chars = sorted({ch for s in train_slice for ch in s if ord(ch) < 128})  # keep ASCII only
# Ensure we include some common URL punctuation if absent
baseline = set(string.ascii_lowercase + string.digits + "/:_-.?=&%")
alphabet = "".join(sorted(set(unique_chars).union(baseline)))

char2idx = {c: i+1 for i, c in enumerate(alphabet)}  # 0 is PAD
vocab_size = len(char2idx) + 1  # +1 for PAD

def encode(series):
    out = np.zeros((len(series), max_len), dtype=np.int32)
    for i, u in enumerate(series.str.lower()):
        seq = [char2idx.get(ch, 0) for ch in u[:max_len]]
        if len(seq) < max_len:
            seq += [0]*(max_len - len(seq))
        out[i] = seq
    return out

X_train = encode(url_train)
X_test  = encode(url_test)

# Sanity check: indices must be < vocab_size
mx = int(X_train.max())
assert mx < vocab_size, f"Max index {mx} >= vocab_size {vocab_size} — mapping/embedding mismatch."

print(f"[CNN] Alphabet size: {len(alphabet)} | vocab_size: {vocab_size} | max index in data: {mx}")
print(f"[CNN] Shapes — train: {X_train.shape}, test: {X_test.shape}")

# --- Build model (no input_length) ---
model = keras.Sequential([
    layers.Embedding(input_dim=vocab_size, output_dim=32),
    layers.Conv1D(64, 5, activation="relu"),
    layers.GlobalMaxPooling1D(),
    layers.Dense(64, activation="relu"),
    layers.Dropout(0.2),
    layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["AUC","Precision","Recall"])
cb = [keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)]

print("[CNN] Training…")
history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=10,
    batch_size=256,
    callbacks=cb,
    verbose=1
)

print("[CNN] Predicting…")
y_prob = model.predict(X_test, verbose=0).ravel()

# --- Threshold that penalizes FN (malicious predicted benign) ---
def pick_threshold_min_cost(y_true, y_prob, fn_weight=5.0):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    candidates = np.r_[thr, 1.0 - 1e-12]
    best_t, best_cost = 0.5, float("inf")
    for t in candidates:
        yhat = (y_prob >= t).astype(int)
        FN = int(((y_true==1)&(yhat==0)).sum())
        FP = int(((y_true==0)&(yhat==1)).sum())
        cost = fn_weight*FN + FP
        if cost < best_cost:
            best_t, best_cost = float(t), cost
    return best_t

thr = pick_threshold_min_cost(y_test.values, y_prob, fn_weight=5)
y_pred = (y_prob >= thr).astype(int)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, zero_division=0)
rec = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)

print(f"\n[CNN] Results @ threshold={thr:.3f}")
print(f"Accuracy:  {acc:.3f}")
print(f"Precision: {prec:.3f}")
print(f"Recall:    {rec:.3f}")
print(f"F1 Score:  {f1:.3f}")

# Save artifacts
model.save("../notebooks/models/url_cnn_charlevel.keras")
pd.DataFrame([{
    "Model": "Char-CNN (URL only)",
    "Threshold": thr,
    "Accuracy": acc,
    "Precision": prec,
    "Recall": rec,
    "F1 Score": f1
}]).to_csv("../data/cnn_results.csv", index=False)
print("[CNN] Saved model and results ✓")
```

```python

# === Load data ===
df = pd.read_csv("../data/df_combined.csv")

# Your selected columns
columns_to_keep = [
    'url','label',
    'num_digits','url_length','num_subdomains','num_path_segments',
    'num_suspicious_keywords','url_entropy','jaccard_hostname_path','path_self_similarity',
    'is_in_alexa','log_ranking','suspicious_subdomain_or_path',
    'domain_word2vec_similarity','avg_token_length','digit_letter_ratio'
]
df = df[columns_to_keep].copy()

# Train/test split (stratified)
X = df.drop(columns=['label','url'])
y = df['label'].astype(int)
X_train, X_test, y_train, y_test, url_train, url_test = train_test_split(
    X, y, df['url'], test_size=0.2, random_state=42, stratify=y
)

from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
# ---------------------- Fit Surrogate Tree ----------------------
# A shallow tree (depth 3-4) yields readable rules.
tree = DecisionTreeClassifier(
    max_depth=4,
    min_samples_leaf=25,
    class_weight="balanced",
    random_state=42
)
tree.fit(X_train, y_train)

# Export text rules
rules_text = export_text(tree, feature_names=list(X.columns), decimals=3)

# Evaluate briefly
report = classification_report(y_test, tree.predict(X_test), output_dict=True, zero_division=0)
metrics_df = pd.DataFrame(report).T

# ---------------------- Build Leaf Rule Table ----------------------
# For each leaf, compute: support, phishing rate (precision on that leaf), and condition string
from sklearn import tree as sktree

node_indicator = tree.decision_path(X)  # on full sample to get leaf coverage
leaf_ids = tree.apply(X)

# Build readable conditions per leaf
def extract_rule_for_leaf(clf, leaf_id, feature_names):
    # Traverse from root to leaf to collect splits
    children_left = clf.tree_.children_left
    children_right = clf.tree_.children_right
    threshold = clf.tree_.threshold
    features = clf.tree_.feature

    # find path
    path = []
    node = leaf_id
    # backtrack: build parents map first
    parents = {}
    stack = [0]
    while stack:
        n = stack.pop()
        left = children_left[n]
        right = children_right[n]
        if left != -1:
            parents[left] = (n, "<=", threshold[n], features[n])
            stack.append(left)
        if right != -1:
            parents[right] = (n, ">", threshold[n], features[n])
            stack.append(right)

    cur = leaf_id
    while cur in parents:
        parent, op, thr, feat_idx = parents[cur]
        feat = feature_names[feat_idx] if feat_idx >= 0 else "N/A"
        if feat != "N/A":
            path.append(f"{feat} {op} {thr:.3f}")
        cur = parent
    path = path[::-1]
    return " AND ".join(path) if path else "Always"

# Collect per-leaf stats
rows = []
unique_leaf_ids = np.unique(leaf_ids)
for lid in unique_leaf_ids:
    mask = (leaf_ids == lid)
    support = int(mask.sum())
    phish = int(y[mask].sum())
    legit = support - phish
    phishing_rate = phish / support if support > 0 else 0.0
    rule = extract_rule_for_leaf(tree, lid, list(X.columns))
    # predicted class at the leaf:
    value = tree.tree_.value[lid][0]
    pred_class = int(np.argmax(value))
    rows.append({
        "LeafID": int(lid),
        "Predicted": pred_class,
        "Support": support,
        "Phish_Count": phish,
        "Legit_Count": legit,
        "Phish_Rate": round(phishing_rate, 3),
        "Rule": rule
    })

rules_df = pd.DataFrame(rows).sort_values(["Predicted","Phish_Rate","Support"], ascending=[False, False, False])

# ---------------------- Save Plot ----------------------
fig_path = Path("/mnt/data/surrogate_tree_rules.png")
plt.figure(figsize=(12, 7))
plot_tree(tree, feature_names=list(X.columns), class_names=["Benign","Phishing"],
          filled=True, impurity=False, proportion=True, rounded=True, fontsize=9)
plt.title("Surrogate Decision Tree (Depth ≤ 4) summarizing dominant URL rules")
plt.tight_layout()
plt.savefig(fig_path, dpi=300)
```

```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, precision_recall_curve

# --- Load data ---
df_train = pd.read_csv("../data/df_combined.csv")
df_tank = pd.read_csv("../data/df_tank_feat.csv")

# --- Columns to use (same as before) ---
cols = [
    'num_digits','url_length','num_subdomains','num_path_segments',
    'num_suspicious_keywords','url_entropy','jaccard_hostname_path','path_self_similarity',
    'is_in_alexa','log_ranking','suspicious_subdomain_or_path',
    'domain_word2vec_similarity','avg_token_length','digit_letter_ratio'
]

X_train = df_train[cols].copy()
y_train = df_train['label'].astype(int)

# Remove url/label from tank and align columns
X_tank = df_tank[cols].copy()
y_tank = df_tank['label'].astype(int) if 'label' in df_tank.columns else np.ones(len(df_tank))  # all phishing

# --- Scale features (fit on training only) ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_tank_scaled = scaler.transform(X_tank)

# --- Helper functions ---
def pick_threshold_min_cost(y_true, y_prob, fn_weight=5.0):
    prec, rec, thr = precision_recall_curve(y_true, y_prob)
    candidates = np.r_[thr, 1.0 - 1e-12]
    best_t, best_cost = 0.5, np.inf
    for t in candidates:
        yhat = (y_prob >= t).astype(int)
        FN = ((y_true==1)&(yhat==0)).sum()
        FP = ((y_true==0)&(yhat==1)).sum()
        cost = fn_weight*FN + FP
        if cost < best_cost:
            best_t, best_cost = float(t), cost
    return best_t

def evaluate(y_true, y_prob, threshold):
    y_pred = (y_prob >= threshold).astype(int)
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, zero_division=0),
        "Recall": recall_score(y_true, y_pred, zero_division=0),
        "F1 Score": f1_score(y_true, y_pred, zero_division=0)
    }

# --- Train models ---
models = {}

# Logistic L2
log_l2 = LogisticRegression(max_iter=5000, solver="lbfgs", penalty="l2")
log_l2.fit(X_train_scaled, y_train)
models["LogReg (L2)"] = log_l2

# Logistic L1
log_l1 = LogisticRegression(max_iter=5000, solver="saga", penalty="l1", C=0.01)
log_l1.fit(X_train_scaled, y_train)
models["LogReg (L1)"] = log_l1

# Random Forest
rf = RandomForestClassifier(n_estimators=400, max_depth=None, min_samples_split=5,
                            class_weight="balanced_subsample", random_state=42, n_jobs=-1)
rf.fit(X_train_scaled, y_train)
models["Random Forest"] = rf

# Linear SVM (calibrated)
svc = LinearSVC(class_weight={0:1,1:3}, max_iter=5000)
cal_svc = CalibratedClassifierCV(svc, cv=5, method="sigmoid")
cal_svc.fit(X_train_scaled, y_train)
models["Linear SVM (Calibrated)"] = cal_svc

# --- Evaluate on PhishTank (external) ---
results = []
for name, model in models.items():
    try:
        y_prob = model.predict_proba(X_tank_scaled)[:,1]
    except AttributeError:
        y_prob = cal_svc.predict_proba(X_tank_scaled)[:,1]
    thr = pick_threshold_min_cost(y_train, model.predict_proba(X_train_scaled)[:,1], fn_weight=5)
    metrics = evaluate(y_tank, y_prob, thr)
    metrics.update({"Model": name, "Threshold": thr})
    results.append(metrics)

res_df = pd.DataFrame(results)[["Model","Accuracy","Precision","Recall","F1 Score","Threshold"]]
res_df = res_df.sort_values(by="Recall", ascending=False).reset_index(drop=True)

# --- Save results ---
out_path = Path("../data/model_results_tank.csv")
res_df.to_csv(out_path, index=False)
print(f"Saved external validation results -> {out_path}")
res_df
```
